```python
# convert the datatypes of START_DATE and END_date columns to datetime 
df.dtypes
```

```python
df["START_DATE"] = pd.to_datetime(df["START_DATE"])
df["END_DATE"] = pd.to_datetime(df["END_DATE"]) 
```

```python
df.dtypes
```

```python
# Extract the MONTh from START_DATE and try to get the proportion of rides of different months .
df['Month'] = pd.DatetimeIndex(df["START_DATE"]).month
```

```python
df.head(2)
```

```python
df['Month_1'] = df['START_DATE'].dt.month
df['Month_Name'] = df['START_DATE'].dt.month_name()
```

```python
df.head(2)
```

```python
df[["Month" , "Month_Name"]].value_counts().reset_index(name='Count') 
```

```python
# Find the average distance covered each month 
df.groupby("Month").mean(numeric_only=True) 
```

```python
# Extract the DAY from the start date column .
df["Day"] = df["START_DATE"].dt.day
df.head(2)
```

```python
#Find the Total Miles covered per category per purpose ;
df.groupby(["CATEGORY" , "PURPOSE"]).sum(numeric_only =True)['MILES'].sort_values(ascending = False).reset_index(name="Total Miles")
```
